```python

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt




from sklearn.model_selection import train_test_split #for training and testing data split
from sklearn.preprocessing import StandardScaler #for feature scaling
from sklearn.linear_model import LogisticRegression #for model building
from sklearn.metrics import classification_report, confusion_matrix #model relation
from sklearn.pipeline import Pipeline #trading 
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.read_csv('/Users/joestar/Desktop/machine learning/AIML Dataset.csv')
```

```python
df.head()


```

```python
df.drop("step",axis=1,inplace=True)

```

```python
df_model=df.drop(['nameOrig','nameDest',"isFlaggedFraud"],axis=1)
```

```python
categorial=df[['type']]
numeric=df[['amount','oldbalanceOrg','newbalanceOrig','oldbalanceDest','newbalanceDest']]
```

```python
y=df_model['isFraud']
x=df_model.drop('isFraud',axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,stratify=y,random_state=42)
```

```python
preprocessor=ColumnTransformer(transformers=[('num',StandardScaler(),numeric.columns),('cat',OneHotEncoder(drop='first'),categorial.columns)
],
remainder='drop')
```

```python
pipeline=Pipeline(steps=[('prep',preprocessor),('clas',LogisticRegression(class_weight='balanced',max_iter=1000))])
```

```python
pipeline.fit(x_train,y_train)
```

```python
y_pred=pipeline.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
pipeline.score(X=x_test,y=y_test)*100
```

```python
import joblib
joblib.dump(pipeline,'fraud_detection_model.pkl')
```
